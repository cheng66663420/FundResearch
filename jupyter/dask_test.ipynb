```python
import dask.dataframe as dd
import pandas as pd
import numpy as np

# 创建示例数据
n = 1000000
dates = pd.date_range('2020-01-01', periods=n, freq='H')
data = {
    'date': dates,
    'value': np.random.rand(n)
}
df = pd.DataFrame(data)

# 将 Pandas DataFrame 转换为 Dask DataFrame
ddf = dd.from_pandas(df, npartitions=10)

# 添加年月列
ddf['year_month'] = ddf['date'].dt.strftime('%Y%m')

# 按年月分区存储为 Parquet 文件
output_path = 'd:/output_data/'
ddf.to_parquet(
    output_path,
    partition_on=['year_month'],
    partition_filename='{year_month}.parquet',  # 自定义文件名格式
    engine='pyarrow',
    compression='snappy'  # Use this instead of write_options
)

print(f"数据已按年月分区存储在 {output_path} 中。")
```
